```python
import sys

def compress(input_file, dict_size):
    # Verifica o tamanho do dicionário
    if dict_size < 9 or dict_size > 16:
        print("O tamanho do dicionário deve estar entre 9 e 16")
        return

    # Cria dicionário inicial com todos os símbolos possíveis
    dictionary = {bytes([i]): i for i in range(256)}
    next_code = 256

    # Abre o arquivo de entrada em modo binário
    with open(input_file, 'rb') as f:
        # Lê todo o arquivo de entrada
        data = f.read()

    # Inicializa variáveis
    current_str = b''
    compressed_data = []

    # Processa os bytes do arquivo de entrada
    for byte in data:
        # Cria uma nova sequência adicionando o byte atual à sequência atual
        new_str = current_str + bytes([byte])
        # Se a nova sequência já está no dicionário, atualiza a sequência atual
        if new_str in dictionary:
            current_str = new_str
        # Caso contrário, adiciona a nova sequência ao dicionário e atualiza a sequência atual
        else:
            compressed_data.append(dictionary[current_str])
            # Se o dicionário atingiu o tamanho máximo, mantém ele estático
            if next_code <  2**dict_size:
                dictionary[new_str] = next_code
                next_code += 1
                
            current_str = bytes([byte])

    # Adiciona a última sequência ao resultado
    if current_str:
        compressed_data.append(dictionary[current_str])

    bin_list = []
    for num in compressed_data:
        bin_num = format(num, f'0{dict_size}b')
        bin_list.append(bin_num)
        
    #bin_list = [int(x, 2) for x in bin_list]

    return bin_list, dictionary
```
